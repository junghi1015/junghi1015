```python
import sys
import numpy as np
import cv2
import pytesseract

src = cv2.imread('./img/42.jpg', cv2.IMREAD_GRAYSCALE)

if src is None:
    print('Image load failed!')
    sys.exit()

cv2.imshow('src', src)

for sigma in range(1, 50):
    blurred = cv2.GaussianBlur(src, (0, 0), sigma)

    alpha = 1.0
    dst = cv2.addWeighted(src, 1 + alpha, blurred, -alpha, 0.0)

    desc = "sigma: %d" % sigma
    cv2.putText(dst, desc, (10, 30), cv2.FONT_HERSHEY_SIMPLEX,
               1.0, 255, 1, cv2.LINE_AA)

    cv2.imshow('dst', dst)
    cv2.waitKey()

print(pytesseract.image_to_string(dst, lang='eng'))
    
cv2.destroyAllWindows()
```

```python
import sys
import numpy as np
import cv2
import pytesseract


src = cv2.imread('./img/hi.jpg', cv2.IMREAD_GRAYSCALE)

if src is None:
    print('Image load failed!')
    sys.exit()

cv2.imshow('src', src)

sigma = 30
blurred = cv2.GaussianBlur(src, (0, 0), sigma)

alpha = 1.0
dst = cv2.addWeighted(src, 1 + alpha, blurred, -alpha, 0.0)

desc = "sigma: %d" % sigma
cv2.putText(dst, desc, (10, 30), cv2.FONT_HERSHEY_SIMPLEX,
            1.0, 255, 1, cv2.LINE_AA)

cv2.imshow('dst', dst)


se = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 3))
dst1 = cv2.erode(dst, se) # 침식
dst2 = cv2.bitwise_not(dst1)

cv2.imshow('dst1', dst1)
cv2.imshow('dst2', dst2)

print('dst : ' + pytesseract.image_to_string(dst, lang='eng'))
print('dst1 : ' + pytesseract.image_to_string(dst1, lang='eng'))
print('dst2 : ' + pytesseract.image_to_string(dst2, lang='eng'))
print('src : ' + pytesseract.image_to_string(src, lang='eng'))


cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2


def contrast2():
    src = cv2.imread('./img/hi.jpg', cv2.IMREAD_GRAYSCALE)

    if src is None:
        print('Image load failed!')
        return

    # dst(x,y) = src(x,y) + (src(x,y) - 128)*alpha  
    alpha = 1.0
    dst = np.clip(src + (src - 128.)*alpha, 0, 255).astype(np.uint8)

    cv2.imshow('src', src)
    cv2.imshow('dst', dst)
    cv2.waitKey()
    cv2.destroyAllWindows()


if __name__ == '__main__':
    contrast2()
```

```python
import numpy as np
import cv2


def calcGrayHist(img):
    channels = [0]
    histSize = [256]
    histRange = [0, 256]

    hist = cv2.calcHist([img], channels, None, histSize, histRange)

    return hist


def getGrayHistImage(hist):
    histMax = np.max(hist)

    imgHist = np.full((100, 256), 255, dtype=np.uint8)
    for x in range(256):
        pt1 = (x, 100)
        pt2 = (x, 100 - int(hist[x, 0] * 100 / histMax))
        cv2.line(imgHist, pt1, pt2, 0)

    return imgHist


def histgoram_stretching():
    src = cv2.imread('./img/hi.jpg', cv2.IMREAD_GRAYSCALE)

    if src is None:
        print('Image load failed!')
        return

    gmin = float(np.min(src))
    gmax = float(np.max(src))

    dst = ((src - gmin) * 255. / (gmax - gmin)).astype(np.uint8)

    cv2.imshow('src', src)
    cv2.imshow('srcHist', getGrayHistImage(calcGrayHist(src)))

    cv2.imshow('dst', dst)
    cv2.imshow('dstHist', getGrayHistImage(calcGrayHist(dst)))

    cv2.waitKey()
    cv2.destroyAllWindows()


def histgoram_equalization():
    src = cv2.imread('./img/hi.jpg', cv2.IMREAD_GRAYSCALE)

    if src is None:
        print('Image load failed!')
        return

    dst = cv2.equalizeHist(src)

    cv2.imshow('src', src)
    cv2.imshow('srcHist', getGrayHistImage(calcGrayHist(src)))

    cv2.imshow('dst', dst)
    cv2.imshow('dstHist', getGrayHistImage(calcGrayHist(dst)))

    cv2.waitKey()
    cv2.destroyAllWindows()


if __name__ == '__main__':
    histgoram_stretching()
    histgoram_equalization()
    
    
    
    
```
